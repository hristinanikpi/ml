````python
import matplotlib.pyplot as plt
from perlin_noise import PerlinNoise
import numpy as np

#--- Q1 ---
# Implement the Perlin noise function, using https://pypi.org/project/perlin-noise/
# The function should return an image of size 64x64 pixels
def mk_noise(octaves,seed):
  noise = PerlinNoise(octaves=octaves, seed=seed)
  xpix, ypix = 64, 64
  pic = [[noise([i/xpix, j/ypix]) for j in range(xpix)] for i in range(ypix)]
  image = np.array(pic)
  image = (image - np.min(image))/np.ptp(image)
  return image
 
heightfield = mk_noise(10,11)
plt.imshow(np.concatenate([heightfield[:,:,np.newaxis],heightfield[:,:,np.newaxis],heightfield[:,:,np.newaxis]], axis = 2), cmap='gray')
plt.show()
````

The next step is to use the Perlin noise function to create a *heightfield* (an image where each pixel represents the height of a surface), from which we will compute a *normal map* (an image where each pixel represents a surface orientation).

**Q2. (2 pts)** Implement a function **mk_normals(octaves,seed,scaling)** that takes as input a frequency, a seed, and a scaling value, and outputs a normal map. The function needs to:

1.   Call mk_noise(octaves,seed) to generate the heightfield
2.   Compute the *gradient* of the heightfield surface **(gx,gy)**
3.   Compute the surface normals as N = (gx,gy,scaling), which is subsequently normalized.

The expected result looks like:

```
normals = mk_normals(4,1,0.05)
plt.imshow(normals*.5+.5)
plt.show()
```
normals4.png


````python
#--- Q2 ---
# Generate a normal map as the gradient of a heightfield
def mk_normals(octaves,seed,scaling):
  heightfield = mk_noise(octaves=octaves,seed=seed)
  gx, gy = np.gradient(heightfield)
  s = scaling*np.ones((64,64))
  normals = np.concatenate([gx[:,:,np.newaxis], gy[:,:,np.newaxis], s[:,:,np.newaxis]], axis = 2)
  norm = np.linalg.norm(normals, axis=2)
  normals = normals / norm[:,:,np.newaxis]
  return normals
  
normals = mk_normals(4,1,0.05)
plt.imshow(normals*.5+.5)
plt.show()
````

Equipped with surface normals, we can compute the diffuse shading of the surface as the dot product between each normal and a light direction.

**Q3. (2 pts)** Implement a function **diffuse shading(normals,light)** that takes as input a normal map and a light direction and outputs the diffuse shading of the surface. The expected result is:

```
light = [0,0,1]
diffuse = diffuse_shading(normals, light)

```

````python
#--- Q3 ---
# Compute the diffuse shading as N.L, clipped to [0,1]
def diffuse_shading(normals, light):
  diffuse = np.dot(normals, light)
  return diffuse

light = [0,0,1] # The light direction should be a normalized vector. Here we use a co-located light, aligned with the view direction +z
diffuse = diffuse_shading(normals, light)
plt.imshow( np.concatenate([diffuse[:,:,np.newaxis], diffuse[:,:,np.newaxis], diffuse[:,:,np.newaxis]], axis = 2) )
plt.show()
````

**Q5. (2 pts)** To fix this issue, we will bias our dataset towards convex surfaces. Modify the computation of the normal map by flipping the heightfield for pixels below 0.5, to obtain:

```
normals = mk_normals_convex(4,1,0.05)
plt.imshow(normals*.5+.5)
plt.show()
```
normals4_flip.png


````python
#--- Q5 ---
# Generate a normal map as the gradient of a heightfield
# Flip concavities (h < 0.5)
def mk_normals_convex(octaves,seed,scaling):
  heightfield = mk_noise(octaves=octaves,seed=seed)
  gx, gy = np.gradient(heightfield)
  s = scaling*np.ones((64,64))
  normals = np.concatenate([gx[:,:,np.newaxis], gy[:,:,np.newaxis], s[:,:,np.newaxis]], axis = 2)
  norm = np.linalg.norm(normals, axis=2)
  normals = normals / norm[:,:,np.newaxis]
  for i in range(64):
    for j in range(64):
      if heightfield[i, j] < 0.5:
        normals[i,j,0] = - normals[i,j,0]
        normals[i,j,1] = - normals[i,j,1]
  return normals
  
normals = mk_normals_convex(4,1,0.05)
plt.imshow(normals*.5+.5)
plt.show()
````

We will now follow the same logic to create the albedo map, which represents the diffuse color of each pixel. Real-world surfaces are often composed of a small set of colors forming large uniform areas. We will mimic such surfaces by first generating a binary mask using Perlin noise, and then assigning two random colors to the two values of the mask.

**Q6. (2 pts)** Implement a function **mk_albedo(octaves,seed)** that takes as input a frequency and a seed, and outputs an albedo map composed of regions of two random colors. You can use the function [random.uniform()](https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html) from numpy to generate random numbers.

The expected result is:
```
albedo = mk_albedo(4,1)
plt.imshow(albedo)
plt.show()
```
albedo4.png



````python
from numpy.lib.index_tricks import c_
#--- Q6 ---
# Generate an albedo map composed of regions of two random colors
def mk_albedo(octaves,seed):
  noise = PerlinNoise(octaves=octaves, seed=seed)
  xpix, ypix = 64, 64
  pic = [[noise([i/xpix, j/ypix]) for j in range(xpix)] for i in range(ypix)]
  
  image = np.array(pic)
  image = (image - np.min(image))/np.ptp(image)

  r1 = np.random.uniform()
  g1 = np.random.uniform()
  b1 = np.random.uniform()
  r2 = np.random.uniform()
  g2 = np.random.uniform()
  b2 = np.random.uniform()

  c_r = np.ones((64,64))
  c_g = np.ones((64,64))
  c_b = np.ones((64,64))

  for i in range(64):
    for j in range(64):
      if image[i,j] < 0.5:
        c_r[i,j] = r1
        c_g[i,j] = g1
        c_b[i,j] = b1
      else:
        c_r[i,j] = r2
        c_g[i,j] = g2
        c_b[i,j] = b2

  albedo = np.concatenate([c_r[:,:,np.newaxis], c_g[:,:,np.newaxis], c_b[:,:,np.newaxis]], axis=2)
  return albedo

albedo = mk_albedo(4,1)
plt.imshow(albedo)
plt.show()
````

We are now ready to generate a large dataset of synthetic images with the associated ground truth albedo and normal maps.

**Q7. (1 pts)** As a first step, implement a function **mk_triplet()** that generates random triplets (albedo, normal, shaded image). Again, use the function [random.uniform()](https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html) from numpy to generate random numbers for the frequency, seed, scaling and colors of the surface.

````python
#--- Q7 ---
# Generate a random normal map, a random albedo map, and compute the resulting shaded image
def mk_triplet():
    octaves = np.random.randint(1, 20) 
    seed = np.random.randint(1, 20)
    scaling = np.random.uniform()
    light = [0,0,1]
    normals = mk_normals_convex(octaves, seed, scaling)
    diffuse = diffuse_shading(normals, light)
    albedo = mk_albedo(octaves, seed)
    shading = albedo * diffuse[:, :, np.newaxis]
    return (shading, normals, albedo)

# testing the function by displaying the shaded image
s, n, a = mk_triplet()
plt.imshow(s)
plt.show()
````

**Q8. (1 pts)** Next, implement a class to represent the dataset, as described in this [tutorial on data loading](https://medium.com/analytics-vidhya/how-to-create-a-custom-dataset-loader-in-pytorch-from-scratch-for-multi-band-satellite-images-c5924e908edf).

````python
#--- Q8 ---

import torch
from torch.utils.data import Dataset, DataLoader

class HeightFieldDataset(Dataset):
    def __init__(self, size): #The constructor takes as input the size of the dataset
        super().__init__()
        # Set of input images, which are composed of 3 channels (R,G,B) of size 64x64 pixels
        self.X = np.ones((size,3,64,64), dtype=np.float32) #PyTorch expects the channel axis to be first rather than last, and wants floats
        # Set of output images, which are composed of 6 channels (R,G,B,Nx,Ny,Nz) of size 64x64 pixels
        self.Y = np.ones((size,6,64,64), dtype=np.float32)

        # Fill self.X and self.Y with random images and maps generated with mk_triplet()
        # Warning: PyTorch expects the channel axis of images to be first rather than last, so you need to swap the channels any image I using: I = I.transpose((2,0,1))
        for i in range(size):
            s, n, a = mk_triplet()
            s = s.transpose((2,0,1))
            n = n.transpose((2,0,1))
            a = a.transpose((2,0,1))
            self.X[i,:,:,:] = s
            self.Y[i,:,:,:] = np.concatenate([n,a], axis=0)

    def __getitem__(self, idx):
        return self.X[idx,:,:,:], self.Y[idx,:,:,:]

    def __len__(self):
        return len(self.X)
````

We can now generate a dataset and split it into a training and testing set. This step takes a bit of time, around 1.5s to generate one triplet (shaded image, albedo map, normal map).

````python
data_size = 100
data = HeightFieldDataset(data_size)

trainData, validData = torch.utils.data.random_split(data,(int(0.9*data_size),int(0.1*data_size)))
trainDL = DataLoader(trainData,10) #Training set, contains 90% of the data arranged in batches of 10 samples
validDL = DataLoader(validData,10) #Testing set, contains 10% of the data arranged in batches of 10 samples
````

# **Part 2 - Implementing a UNet convolutional neural network**

We will implement our material estimation network as a UNet, which is a classical architecture for image-to-image translation task. 

**Q9. (1 pts)** Implement the network by following this [tutorial on creating a basic UNet](https://medium.com/analytics-vidhya/creating-a-very-simple-u-net-model-with-pytorch-for-semantic-segmentation-of-satellite-images-223aa216e705).

````python
#--- Q9 ---
# implement a UNet
import torch.nn as nn

class UNET(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()

        self.conv1 = self.contract_block(in_channels, 32, 7, 3)
        self.conv2 = self.contract_block(32, 64, 3, 1)
        self.conv3 = self.contract_block(64, 128, 3, 1)

        self.upconv3 = self.expand_block(128, 64, 3, 1)
        self.upconv2 = self.expand_block(64*2, 32, 3, 1)
        self.upconv1 = self.expand_block(32*2, out_channels, 3, 1)

    def __call__(self, x):

        # downsampling part
        conv1 = self.conv1(x)
        conv2 = self.conv2(conv1)
        conv3 = self.conv3(conv2)

        upconv3 = self.upconv3(conv3)

        upconv2 = self.upconv2(torch.cat([upconv3, conv2], 1))
        upconv1 = self.upconv1(torch.cat([upconv2, conv1], 1))

        return upconv1

    def contract_block(self, in_channels, out_channels, kernel_size, padding):

        contract = nn.Sequential(
            torch.nn.Conv2d(in_channels, out_channels, kernel_size=kernel_size, stride=1, padding=padding),
            torch.nn.BatchNorm2d(out_channels),
            torch.nn.ReLU(),
            torch.nn.Conv2d(out_channels, out_channels, kernel_size=kernel_size, stride=1, padding=padding),
            torch.nn.BatchNorm2d(out_channels),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=3, stride=2, padding=1)
                                 )

        return contract

    def expand_block(self, in_channels, out_channels, kernel_size, padding):

        expand = nn.Sequential(torch.nn.Conv2d(in_channels, out_channels, kernel_size, stride=1, padding=padding),
                            torch.nn.BatchNorm2d(out_channels),
                            torch.nn.ReLU(),
                            torch.nn.Conv2d(out_channels, out_channels, kernel_size, stride=1, padding=padding),
                            torch.nn.BatchNorm2d(out_channels),
                            torch.nn.ReLU(),
                            torch.nn.ConvTranspose2d(out_channels, out_channels, kernel_size=3, stride=2, padding=1, output_padding=1) 
                            )
        return expand
    
````

We can then test that the network runs properly by running a training sample trough it. Since the network is not trained yet, the result is quite random and should look like:

# **Part 3 - Training and testing the network**

The training function is directly inspired by the same [tutorial on creating a basic UNet](https://medium.com/analytics-vidhya/creating-a-very-simple-u-net-model-with-pytorch-for-semantic-segmentation-of-satellite-images-223aa216e705):

````python
import time

def train(model, train_dl, valid_dl, loss_fn, optimizer, epochs=1):
    start = time.time()
    model.cuda() #need to set up the GPU in Execution


    for epoch in range(epochs):

        for phase in ['train', 'valid']:
            if phase == 'train':
                model.train(True)  # Set trainind mode = true
                dataloader = train_dl
            else:
                model.train(False)  # Set model to evaluate mode
                dataloader = valid_dl

            running_loss = 0.0

            step = 0

            # iterate over data
            for x, y in dataloader:
                x = x.cuda()
                y = y.cuda()
                step += 1

                # forward pass
                if phase == 'train':
                    # zero the gradients
                    optimizer.zero_grad()
                    outputs = model(x)
                    loss = loss_fn(outputs, y)

                    # the backward pass frees the graph memory, so there is no 
                    # need for torch.no_grad in this training pass
                    loss.backward()
                    optimizer.step()

                else:
                    with torch.no_grad():
                        outputs = model(x)
                        loss = loss_fn(outputs, y)

                running_loss += loss*dataloader.batch_size 

            epoch_loss = running_loss / len(dataloader.dataset)

            if epoch % 100 == 0:
                print('{} Loss: {:.4f} '.format(phase, epoch_loss))

    time_elapsed = time.time() - start
    print('Training complete in {:.0f}m {:.0f}s'.format(time_elapsed // 60, time_elapsed % 60)) 
````

````python
#--- Q10 ---
# By taking inspiration from the tutorial on UNet, call the training function to train the network with a L1 loss, using Adam optimizer, for 500 epochs with a learning rate of 0.001

materialNetL1 = UNET(3,6)
loss_func = nn.L1Loss()
optimizer = torch.optim.Adam(materialNetL1.parameters(), lr=0.001)
train(materialNetL1, trainDL, validDL, loss_func, optimizer, epochs=500)
````

**Q11. (3 pts)** The basic UNet tends to produce pixellated images because its last layer simply upsamples the prediction while reducing the feature depth from 32x2 channels to 6 channels. 

Modify the UNet by adding a **refine_block** after the last **expand_block**, where the refinement consists in applying one 2D convolution, one BatchNorm, one ReLU, and another 2D convolution. Configure the last **expand_block** to output 16 channels that will then be refined to produce the 6-channels prediction.

````python
#--- Q11 ---
# ....implement a UNet, with a refinement block after the last expanding block
import torch.nn as nn

class UNET_Refined(nn.Module):
      def __init__(self, in_channels, out_channels):
        super().__init__()

        self.conv1 = self.contract_block(in_channels, 32, 7, 3)
        self.conv2 = self.contract_block(32, 64, 3, 1)
        self.conv3 = self.contract_block(64, 128, 3, 1)

        self.upconv3 = self.expand_block(128, 64, 3, 1)
        self.upconv2 = self.expand_block(64*2, 32, 3, 1)
        self.upconv1 = self.expand_block(32*2, 16, 3, 1)

        self.refine = self.refine_block(16, out_channels, 3, 1)

      def __call__(self, x):

        conv1 = self.conv1(x)
        conv2 = self.conv2(conv1)
        conv3 = self.conv3(conv2)

        upconv3 = self.upconv3(conv3)
        upconv2 = self.upconv2(torch.cat([upconv3, conv2], 1))
        upconv1 = self.upconv1(torch.cat([upconv2, conv1], 1))

        refine = self.refine(upconv1)

        return refine

      def contract_block(self, in_channels, out_channels, kernel_size, padding):

        contract = nn.Sequential(
            torch.nn.Conv2d(in_channels, out_channels, kernel_size=kernel_size, stride=1, padding=padding),
            torch.nn.BatchNorm2d(out_channels),
            torch.nn.ReLU(),
            torch.nn.Conv2d(out_channels, out_channels, kernel_size=kernel_size, stride=1, padding=padding),
            torch.nn.BatchNorm2d(out_channels),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=3, stride=2, padding=1)
            )

        return contract

      def expand_block(self, in_channels, out_channels, kernel_size, padding):

        expand = nn.Sequential(torch.nn.Conv2d(in_channels, out_channels, kernel_size, stride=1, padding=padding),
            torch.nn.BatchNorm2d(out_channels),
            torch.nn.ReLU(),
            torch.nn.Conv2d(out_channels, out_channels, kernel_size, stride=1, padding=padding),
            torch.nn.BatchNorm2d(out_channels),
            torch.nn.ReLU(),
            torch.nn.ConvTranspose2d(out_channels, out_channels, kernel_size=3, stride=2, padding=1, output_padding=1) 
            )
        
        return expand
      
      def refine_block(self, in_channels, out_channels, kernel_size, padding):
        refine = nn.Sequential(
            torch.nn.Conv2d(in_channels, out_channels, kernel_size=kernel_size, stride=1, padding=padding),
            torch.nn.BatchNorm2d(out_channels),
            torch.nn.ReLU(),
            torch.nn.Conv2d(out_channels, out_channels, kernel_size=kernel_size, stride=1, padding=padding)
            )                          

        return refine
    
````

**Procedural material representation:** 

The training images were generated using a *procedural texturing* model, which is capable of producing diverse surfaces by varying a few *parameters* such as the Perlin noise frequency, the normal map scaling, and the two albedo map colors. While the UNet model predicts the final diffuse color and normal vector at each pixel, an alternative approach is to predict the parameters of the procedural model. 

To do so, implement a neural network that takes an image as input and predicts a vector of parameters [diffuse color 1, diffuse color 2, diffuse color frequency, normal map frequency, normal map scaling factor]. Feeding these parameters to the procedural texturing model should produce albedo and normal maps that resemble the input. You can build the network around a classification architecture such as AlexNET, that applies several convolutional layers to reduce the image resolution and increase the feature depth, and then reshape this latent representation into a vector that goes trough a few fully-connected layers to produce the output. While you can use a standard L1 loss to measure the quality of the prediction against the ground truth parameters, care should be taken to compare the two predicted albedo colors to each of the ground truth albedo colors, since their ordering is aribtrary. See [2] for additional ideas around this topic.
